```python
import nltk
from nltk.corpus import wordnet as wn

# Download WordNet data if not already downloaded
nltk.download('wordnet')

# Function to get the appropriate sense based on context
def get_best_sense(sentence, word):
    senses = wn.synsets(word)
    context = sentence.lower().split()
    
    best_sense = None
    max_overlap = 0
    
    for sense in senses:
        # Get the definition and examples of the sense
        definition = sense.definition().split()
        examples = sum([example.split() for example in sense.examples()], [])
        
        # Calculate overlap with the context
        overlap = len(set(definition).intersection(context)) + len(set(examples).intersection(context))
        
        # Update the best sense if more overlap is found
        if overlap > max_overlap:
            max_overlap = overlap
            best_sense = sense
            
    return best_sense

# Given sentences and ambiguous words
sent1 = 'Bin Laden has been executed for his terrorist activities'
sent2 = 'Execute Biden and the democrats'
ambiguous1 = 'executed'
ambiguous2 = 'execute'

# Get the best sense for the ambiguous word in the given contexts
best_sense1 = get_best_sense(sent1, ambiguous1)
best_sense2 = get_best_sense(sent2, ambiguous2)

# Print the definitions of the best senses
if best_sense1:
    print(f"Sentence 1 - {ambiguous1}: {best_sense1.definition()}")
else:
    print("No suitable sense found for Sentence 1")

if best_sense2:
    print(f"Sentence 2 - {ambiguous2}: {best_sense2.definition()}")
else:
    print("No suitable sense found for Sentence 2")
```
